```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.autograd import Variable
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
mu = [4,4]
sigma = [[2,1],[1,2]]
data = np.random.multivariate_normal(mu,sigma, 500)
df1 = pd.DataFrame(data, columns = ['x','y'])
df1['set'] = ['first']*df1.shape[0]
```

```python
mu = [-4,-4]
sigma = [[2,1],[1,2]]
data = np.random.multivariate_normal(mu,sigma, 500)
df2 = pd.DataFrame(data, columns = ['x','y'])
df2['set'] = ['second']*df2.shape[0]
```

```python
df = pd.concat([df1,df2]).reset_index(drop=True)
# df
```

```python
fig = px.scatter(x = df['x'], y = df['y'], color=df['set'])
fig.show()
```

```python
class GaussianNet(nn.Module):
    
    def __init__(self):
        super().__init__()
        # define the functions you pass through - note that these are pointers to functions!
        self.linear1 = nn.Linear(2,2)  
    
    def forward(self,x):
        out = self.linear1(x)
        return out
```

```python
net = GaussianNet()
print('Linear Weights for First Layer\n',net.linear1.weight)
print('\nBias for First Layer\n',net.linear1.bias)
```

```python
net.forward(df.loc[:,['x','y']])
```
